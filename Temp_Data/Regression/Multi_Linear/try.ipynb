```python
import pandas as pd  
import numpy as np 
from scipy import stats
import matplotlib.pyplot as plt  
# import seaborn as seabornInstance 
# from sklearn.model_selection import train_test_split 
# from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
import statsmodels.formula.api as smf
```

```python
airbnb_df = pd.read_csv("AirBnB_Reviews_Zipcode.csv")
census_df = pd.read_csv("housing_census_data.csv")
```

```python
housing_census_df = pd.merge(census_df, airbnb_df, on="Zipcode")
housing_census_df.head()
```

```python
housing_income_df = housing_census_df[["Median Rent", "Poverty Count"]]
housing_income_df.head()
housing_income_df.corr(method='pearson')
```

```python
rent = housing_census_df["Median Rent"]
income = housing_census_df["Household Income"]
poverty = housing_census_df["Poverty Rate"]
reviews = housing_census_df["Review Count"]
```

```python
lm = smf.ols(formula='rent ~ income + poverty + reviews', data=housing_census_df).fit()
lm.summary()
```
