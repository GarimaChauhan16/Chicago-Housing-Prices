```python
# Dependencies
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
```

```python
housing_census_df=pd.read_csv("housing_census_data.csv")
housing_census_df.head()
```

```python
median_rent = housing_census_df.iloc[:, 1]
house_income = housing_census_df.iloc[:, 5]
capita_income = housing_census_df.iloc[:, 6]
poverty_rate = housing_census_df.iloc[:, 8]
```

```python
h_income_slope, h_income_int, h_income_r, h_income_p, h_income_std_err = stats.linregress(house_income, median_rent)
h_income_rent_fit = h_income_slope * house_income + h_income_int
```

```python
# plot rent vs household income
fig,axl = plt.subplots()
axl.set_facecolor('lavender')
plt.scatter(housing_census_df['Household Income'], housing_census_df['Median Rent'], edgecolor = 'black', marker='o')
plt.plot(house_income, h_income_rent_fit, linewidth=1, color = "red")
plt.grid (color='white')
plt.title("Median Rent vs. Household Income by Zipcode", fontsize=14)
plt.xlabel("Household Income ($)", fontsize=12)
plt.ylabel("Median Rent ($)", fontsize=12)
# Save Figure as png
plt.savefig("../../Images/Rent_Vs_Household_income_regression.png", bbox_inches='tight')
plt.show()
```

```python
c_income_slope, c_income_int, c_income_r, c_income_p, c_income_std_err = stats.linregress(capita_income, median_rent)
c_income_rent_fit = c_income_slope * capita_income + c_income_int
```

```python
# plot rent vs Per Capita income
fig,axl = plt.subplots()
axl.set_facecolor('lavender')
plt.scatter(housing_census_df['Per Capita Income'], housing_census_df['Median Rent'], edgecolor = 'black', marker='o')
plt.plot(capita_income, c_income_rent_fit, linewidth=1, color = "red")
plt.grid (color='white')
plt.title("Median Rent vs. Per Capita Income by Zipcode", fontsize=14)
plt.xlabel("Per Capita Income ($)", fontsize=12)
plt.ylabel("Median Rent ($)", fontsize=12)
# Save Figure as png
plt.savefig("../../Images/Rent_Vs_PerCapita_income_regression.png", bbox_inches='tight')
plt.show()
```

```python
poverty_slope, poverty_int, poverty_r, poverty_p, poverty_std_err = stats.linregress(poverty_rate, median_rent)
poverty_rent_fit = poverty_slope * poverty_rate + poverty_int
```

```python
# plot rent vs Poverty Rate
fig,axl = plt.subplots()
axl.set_facecolor('lavender')
plt.scatter(housing_census_df['Poverty Rate'], housing_census_df['Median Rent'], edgecolor = 'black', marker='o')
plt.plot(poverty_rate, poverty_rent_fit, linewidth=1, color = "red")
plt.grid (color='white')
plt.title("Median Rent vs. Poverty Rate by Zipcode", fontsize=14)
plt.xlabel("Poverty Rate", fontsize=12)
plt.ylabel("Median Rent ($)", fontsize=12)
# Save Figure as png
plt.savefig("../../Images/Rent_Vs_PovertyRate_regression.png", bbox_inches='tight')
plt.show()
```
